```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
%matplotlib inline
```

loading dataset


```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arunjangir245/boston-housing-dataset")

print("Path to dataset files:", path)

```

```python
import os

# List files in the downloaded directory to find the CSV file
file_list = os.listdir(path)
print("Files in the dataset directory:", file_list)

# Assuming the CSV file is named 'BostonHousing.csv' based on common usage,
# you might need to adjust this based on the actual file list printed above.
csv_file_path = os.path.join(path, 'BostonHousing.csv')

df = pd.read_csv(csv_file_path)
display(df.head())
```

```python
df.describe()
```

description of the dataset

```python
X  = df.drop('medv',axis=1)
y = df['medv']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
# importing base estimators for the ensemble learning like linear regression and knn regressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
LR  = LinearRegression()
KNN = KNeighborsRegressor()
```

```python
LR.fit(X_train,y_train)
KNN.fit(X_train,y_train)
```

```python
# Impute missing values in the 'rm' column with the mean
df['rm'].fillna(df['rm'].mean(), inplace=True)

# Verify that there are no more missing values
print(df.isnull().sum())
```

```python
y_pred = LR.predict(X_test)
y_pred1 = KNN.predict(X_test)
```

```python
y_train = LR.predict(X_train)
y_train1 = KNN.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f"KNN Mean Squared Error: {mean_squared_error(y_test, y_pred1)}")
print(f"KNN R-squared: {r2_score(y_test, y_pred1)}")
print(f"LR Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"LR R-squared: {r2_score(y_test, y_pred)}")
```

```python
bag_reg = BaggingRegressor(random_state=42)
bag_reg.fit(X_train,y_train)
```

```python
y_pred = bag_reg.predict(X_test)
print(f"Bagging Mean Squared Error: {mean_squared_error(y_test, y_pred)}")
print(f"Bagging R-squared: {r2_score(y_test, y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV
n_samples , n_features = X.shape
params = {'estimator': [None, LinearRegression(), KNeighborsRegressor()],
          'n_estimators': [20,50,100],
          'max_samples': [0.5,1.0],
          'max_features': [0.5,1.0],
          'bootstrap': [True, False],
          'bootstrap_features': [True, False]}

bagging_regressor_grid = GridSearchCV(BaggingRegressor(random_state=1, n_jobs=-1), param_grid =params, cv=3, n_jobs=-1, verbose=1)
bagging_regressor_grid.fit(X_train, y_train)

print('Train R^2 Score : %.3f'%bagging_regressor_grid.best_estimator_.score(X_train, y_train))
print('Test R^2 Score : %.3f'%bagging_regressor_grid.best_estimator_.score(X_test, y_test))
print('Best R^2 Score Through Grid Search : %.3f'%bagging_regressor_grid.best_score_)
print('Best Parameters : ',bagging_regressor_grid.best_params_)
```

```python
bag_tree_updated = BaggingRegressor(estimator=LinearRegression(),n_estimators=100,max_samples=0.5,bootstrap=True,max_features=1.0,bootstrap_features=False,random_state=42)
bag_tree_updated.fit(X_train,y_train)
```

```python
y_pred_updated = bag_tree_updated.predict(X_test)
print(f"Bagging Mean Squared Error: {mean_squared_error(y_test, y_pred_updated)}")
print(f"Bagging R-squared: {r2_score(y_test, y_pred_updated)}")
```
